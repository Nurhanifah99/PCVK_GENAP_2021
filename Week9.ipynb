```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import math
from google.colab.patches import cv2_imshow
from PIL import Image as im
import os
```

```python
original = cv.imread('/content/drive/MyDrive/PCVK_GENAP2021/Images/j.png')
img_gray = cv.cvtColor(original,cv.COLOR_BGR2GRAY)

cv2_imshow(img_gray)
```

```python
kernel = np.ones((5,5), np.uint8)
dilation = cv.dilate(img_gray, kernel, iterations = 1)

plt.subplot(121),plt.imshow(img_gray),plt.title('Citra Awal')
plt.subplot(122),plt.imshow(dilation),plt.title('Citra Hasil Dilasi')
```

```python
kernel = np.ones((5,5), np.uint8)
erosion = cv.morphologyEx(img_gray, cv.MORPH_ERODE, kernel)

plt.subplot(121), plt.imshow(img_gray), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(erosion), plt.title('Citra Hasil Erosi')
```

```python
kernel = np.ones((5,5), np.uint8)
dilation = cv.morphologyEx(img_gray, cv.MORPH_DILATE, kernel)

plt.subplot(121), plt.imshow(img_gray), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(dilation), plt.title('Citra Hasil Dilasi')
```

```python
kernel = np.ones((5,5), np.uint8)
erosion = cv.morphologyEx(img_gray, cv.MORPH_ERODE, kernel)

plt.subplot(121), plt.imshow(img_gray), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(erosion), plt.title('Citra Hasil Erosi')
```

```python
kernel = np.ones((7,7), np.uint8)
closing = cv.morphologyEx(img_gray, cv.MORPH_CLOSE, kernel)

plt.subplot(121), plt.imshow(img_gray), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(closing), plt.title('Citra Hasil Closing')
```

```python
#Nomer 9 (1)

kernel = cv.getStructuringElement(cv.MORPH_CROSS, (3, 3))
cross3 = cv.morphologyEx(img_gray, cv.MORPH_DILATE, kernel)

plt.subplot(121), plt.imshow(img_gray), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(cross3), plt.title('Citra Hasil Dilasi 3x3')
```

```python
element = cv.getStructuringElement(cv.MORPH_RECT,(5,5))
dilate_dst = cv.dilate(img_gray, element)
closing_dst = cv.erode(dilate_dst, element)

plt.subplot(121), plt.imshow(img_gray), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(closing_dst), plt.title('Citra Hasil 5,5')
```

```python
element = cv.getStructuringElement(cv.MORPH_RECT,(3,3))
dilate_dst = cv.dilate(img_gray, element)
closing_dst = cv.erode(dilate_dst, element)

plt.subplot(121), plt.imshow(img_gray), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(closing_dst), plt.title('Citra Hasil 3x3')
```

```python
kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (3, 5))
cross3 = cv.morphologyEx(img_gray, cv.MORPH_DILATE, kernel)

plt.subplot(121), plt.imshow(img_gray), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(cross3), plt.title('Citra Hasil Circular 3x5')
```
