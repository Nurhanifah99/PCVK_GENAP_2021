```
# No.5
import random

blank = np.zeros(shape=(512,512,3),dtype=np.int32)

#kotak putih
k1_b = random.randint(0,511)
k1_k = random.randint(0,511)
kotak1 = cv.rectangle(blank, pt1=(k1_b,k1_b), pt2=(k1_k,k1_k), color=(255,255,255),thickness=10)

#kotak cyan
k2_b = random.randint(0,511)
k2_k = random.randint(0,511)
kotak1 = cv.rectangle(blank, pt1=(k2_b,k2_b), pt2=(k2_k,k2_k), color=(0,255,255),thickness=10)

#kotak ungu
k3_b = random.randint(0,511)
k3_k = random.randint(0,511)
kotak1 = cv.rectangle(blank, pt1=(k3_b,k3_b), pt2=(k3_k,k3_k), color=(255,0,255),thickness=10)

#kotak kuning
k4_b = random.randint(0,511)
k4_k = random.randint(0,511)
kotak1 = cv.rectangle(blank, pt1=(k4_b,k4_b), pt2=(k4_k,k4_k), color=(255,255,0),thickness=10)

#kotak 
k5_b = random.randint(0,511)
k5_k = random.randint(0,511)
kotak1 = cv.rectangle(blank, pt1=(k5_b,k5_b), pt2=(k5_k,k5_k), color=(0,0,255),thickness=10)

plt.imshow(kotak1)
```
